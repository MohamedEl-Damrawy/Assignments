```python
# 7. Try aliasing vs cloning with a list and modify one. 
# 1- aliasing 
x = [1,2,3,4]
print(f"x before aliasing  >> {x}")
y = x
y.append(5)
print(f"x after aliasing   >> {x}")

# 2- cloning
a = [10,15,20,25]
print(f"a before cloning  >> {a}")
b = a.copy()
b.append(50)
print(f"a after cloning   >> {a}")



```
